```python
import sys

import numpy as np

from Bio import AlignIO
from Bio.Cluster import kcluster
from Bio import SeqIO
# Read the alignment file and print it to the screen

from Bio.Phylo.TreeConstruction import DistanceTreeConstructor
from Bio.Phylo.TreeConstruction import DistanceCalculator
import matplotlib.pyplot as plt


np.set_printoptions(threshold=sys.maxsize)

# print every line of the matrix on one line
np.set_printoptions(linewidth=np.inf)

# make the output to scroll horizontally so we can have a nicely formatted score matrix using the above 2 numpy functions
from IPython.display import display, HTML
display(HTML("<style>pre { white-space: pre !important; }</style>"))

# Open the FASTA file of the protein sequences
proteinSequences= []
with open("proteins/adenosine_BLAST_sequences.txt", "r") as handle:
    # Use the parse function of SeqIO to iterate over the records in the file
    for record in SeqIO.parse(handle, "fasta"):
        # Each record is a SeqRecord object that contains the sequence data
        
        proteinSequences.append(record.description.split())

print(proteinSequences)


# Read the alignment file and print it to the screen
alignment = AlignIO.read("proteins/adenosine_msa_output", "fasta")
print(alignment)

#replace the name of the organism to the sequence ID (Taxa)
for record in alignment:
    for items in proteinSequences:
        if record.id == items[0]:
            record.id = items[-2] + "_" + items[-1]

#write the edited aligned file in a fasta file
AlignIO.write(alignment,"proteins/adenosine_msa_output_edited","fasta")
print(alignment)
        

    
sequences = [seq.seq for seq in alignment]
db = []
for i in sequences:
    db.append(str(i))
print("Sequences as a list of strings: ")
print(db)
print("\n")

matrix = np.asarray([np.fromstring(s, dtype=np.uint8) for s in db])
print("Sequences as a 2 D array of integers:")
print(matrix)
print("\n")
clusterid, error, nfound = kcluster(matrix, nclusters=2, mask=None, weight=None, transpose=False, npass=1000, method='a', dist='e', initialid=None)
print("Cluster IDs for each sequence:")
print(clusterid)
print("\n")

#writing clusters into a fasta format files
cluster1 = []
cluster2 = []


for i in range(len(clusterid)):
    
    
    if clusterid[i]==1:
        cluster1.append(alignment[i])
        
    else:
        cluster2.append(alignment[i])
        

SeqIO.write(cluster1, "proteins/adenosine_cluster1", "fasta")
SeqIO.write(cluster2, "proteins/adenosine_cluster2", "fasta")
```

```python
plt.rcParams['figure.figsize'] = [15, 4.8]
aln = AlignIO.read('proteins/adenosine_msa_output_edited', 'fasta')
constructor = DistanceTreeConstructor()
calculator = DistanceCalculator('identity')
dm = calculator.get_distance(aln)
upgmatree = constructor.upgma(dm)
Phylo.draw(upgmatree)

#saving the tree of group of proteins into a file in newick format
with open('trees.dnd', 'w') as f:
    Phylo.write(upgmatree, "trees.dnd", "newick")
f.close()
```
